```python
import os 
import numpy as np
import pandas as pd
import json
```

```python
name_folder = "../ressources/"
```

```python
list_contain_folder  = os.listdir(name_folder)
```

```python
list_contain_folder
```

```python
question = []
idQuestion = []
response = []
label = []
for _ in list_contain_folder:
    if 'txt' in _:
         with open(name_folder + f"{_}", "r") as file:
            i = 1
            for line in file:
                if '?' in line:
                    question.append(line.strip())
                    idQuestion.append(_[:3] + f"q{i}")
                    label.append(_.split('_')[0])
                    i +=1
                else:
                    if line :
                        response.append(line.strip())
                    else:
                        print("blank line")




dataset = {
    "idQuestion" : idQuestion,
    "question" : question,
    "answer" : response,
    "label" : label
}
df = pd.DataFrame(dataset)
```

```python
len(question), len(idQuestion), len(response), len(label)
```

```python
df.head()
```

```python
# Saving on CSV file
df.to_csv("../datasets/faq_dataset.csv", index=False)
```

```python
result = df.to_json(orient="records")
parsed = json.loads(result)
format_json = json.dumps(parsed, indent=4)
```

```python
# Afficher le JSON formaté
print(format_json)
```

```python
# Saving
with open('../datasets/faq_dataset.json', 'w') as f:
    f.write(format_json)
```
