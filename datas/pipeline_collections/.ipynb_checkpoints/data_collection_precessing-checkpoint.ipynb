```python
import os 
import numpy as np
import pandas as pd
```

```python
name_folder = "./datas/ressources/"
```

```python
list_contain_folder  = os.listdir(name_folder)
```

```python
list_contain_folder
```

```python
for _ in list_contain_folder:
    if 'txt' in _:
        with open(name_folder + f"{_}", "r") as file:
            for line in file:
                print(line)
```

```python
question = []
idQuestion = []
response = []
label = []
for _ in list_contain_folder:
    if 'txt' in _:
         with open(name_folder + f"{_}", "r") as file:
            i = 1
            for line in file:
                if '?' in line:
                    question.append(line.strip())
                    idQuestion.append(_[:3] + f"q{i}")
                    label.append(_.split('_')[0])
                    i +=1
                else:
                    if line :
                        response.append(line.strip())
                    else:
                        print("blank line")




dataset = {
    "idQuestion" : idQuestion,
    "question" : question,
    "answer" : response,
    "label" : label
}
df = pd.DataFrame(dataset)
```

```python
# Saving on CSV file
df.to_csv("../datas/datasets/faq_dataset.csv", index=False)
```
